```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Loading model

```python
cnn=tf.keras.models.load_model('trained_model.h5')
```

# visualization and performing prediction on single image

```python
import cv2
image_path='test/capsicum/Image_1.jpg'
img=cv2.imread(image_path)
plt.imshow(img)
plt.title('Test_Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#testing model
```

```python
image=tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr=tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])
prediction=cnn.predict(input_arr)
```

```python
print(prediction)
```

```python
test_set=tf.keras.utils.image_dataset_from_directory(
"test",labels='inferred',
label_mode='categorical',
class_names=None,
color_mode='rgb',
batch_size=32,
image_size=(64,64),
shuffle=True,
seed=None,
validation_split=None,
subset=None,
interpolation='bilinear',
follow_links=False,
crop_to_aspect_ratio=False)
```

```python
test_set.class_names
```

```python
result_index=np.where(prediction[0]==max(prediction[0]))
```

```python
print(result_index[0][0])
```

```python
#Display image
plt.imshow(img)
plt.title('Test_Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

# single prediction

```python
print("it's a {}".format(test_set.class_names[result_index[0][0]]))
```
